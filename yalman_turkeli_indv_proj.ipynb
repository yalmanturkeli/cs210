```python
print("2) Create two new columns named ”pickup district” and ”dropoff district” by applying ")
print("   reverse geocoding to associated coordinates.")
print(" ")
print("Pickup District:")

lst = [] 

for i, entry in df.iterrows():
  longitude = entry["pickup_longitude"]
  latitude = entry["pickup_latitude"]
  coord = latitude, longitude
  lst.append(coord)


results = rg.search(lst)

dist=[]
for j in range(len(results)):
  dist.append(results[j]['name']+", " +results[j]['admin1'])

  
df["pickup_district"]=dist

df.head() 
```

```python
print("Dropoff District:")

lst2 = [] 

for i, entry in df.iterrows():
  longitude2 = entry["dropoff_longitude"]
  latitude2 = entry["dropoff_latitude"]
  coord2 = latitude2, longitude2
  lst2.append(coord2)


results2 = rg.search(lst2)

dist2=[]
for j in range(len(results2)):
  dist2.append(results2[j]['name']+", " +results2[j]['admin1'])

  
df["dropoff_district"]=dist2

df.head()
```

```python
print("3) Extract the top 5 districts where passengers prefer to leave and arrive.")
print(" ")
print("For leave:")

tops = df.groupby("pickup_district").count().sort_values(by='id', ascending=False)

tops.head(5)
```

```python
print("For arrive:")

tops = df.groupby("dropoff_district").count().sort_values(by='id', ascending=False)

tops.head(5)
```

```python
print("4) Create a new column named ”distance” by utilizing pick up and drop off coordinates. ")

coordinates = [] 
coordinates2=[]
lstc=[]
for i,entry in df.iterrows():
  longitude=entry["pickup_longitude"]
  latitude=entry["pickup_latitude"]
  coord= latitude, longitude
  coordinates.append(coord)
  longitude2= entry["pickup_longitude"]
  latitude2=entry["pickup_latitude"]
  coord2= longitude2,  latitude2 
  coordinates2.append(coord2)

for x in range(len(coordinates2)):
  lstc.append(geodesic(coordinates[x], coordinates2[x]).miles)
  

df["distance"]=lstc

df.head()
```

```python
print("5) Create a new column named ”time of day” by aggregating timestamps in ”pickup datetime”")
print("   into 5 different categories.")

time_day=[]

for i,entry in df.iterrows():
  time= entry["pickup_datetime"]
  t= int(time[11:13])
  if t>=7 and t<9:
    time_day.append("rush hour morning") 
  elif t>=9 and t<16:
    time_day.append("afternoon")
  elif t>=16 and t<18:
    time_day.append("rush hour evening")
  elif t>=18 and t<23:
    time_day.append("evening")  
  elif t>=23 or t<7:
    time_day.append("late night")  
  
  

df["time_of_day"]= time_day
df.head()
```

```python
print("6) Show how the average distance varies as time of the day changes.")
print("Average distance has not changes much related to time of days so it does not affect it.")
print(" ")
print("7) Show how the trip duration varies as time of the day changes.")
print("In the trip durtion, evening and late nigth is smallest but rush hours evening hightest.")


df.groupby(by="time_of_day").mean()
```
